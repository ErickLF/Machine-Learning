## Generacion de numeros aleatorios

```python
import numpy as np
```

```python
#Genera un numero aleatorio entre 1 y 100
np.random.randint(1,100)
```

```python
#La forma mas clasica de generar un numero aleatorio es ente 0 y 1
np.random.random()
```

```python
#Funcion que genera una lista de n numeros aleatorios dentro del intervalo [a,b]
def randIntList(n,a,b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randIntList(10,3,60)
```
