# Carga de datos a traves de la función open
Ideal para carga de datos muy muy grandes (ejemplo: 15 - 20 GB)
- Lo que hace la funcion open es leer linea por linea ejecutando un ciclo

```python
#Leemos el churn model
data3 = open(os.path.join(mainpath,directorio),"r")
```

```python
#Extraemos la informacion linea a lines
cols = data3.readline().strip().split(",")
n_cols = len(cols)
n_cols
```

```python
#script para contar las filas
counter = 0
main_dict = {}
for col in cols:
    main_dict[col]= []#nombre de la columna como clave
```

```python
main_dict
```

```python
#iteramos sobre cada una de las filas del data set
for line in data3:
    values = line.strip().split(",")
    #iteramos sobre el renglon para obtener todas las columnas
    for i in range(len(cols)):
        #para cada columna del data set se convierte clave del diccionario
        main_dict[cols[i]].append(values[i])
    counter +=1
print("El data set tiene %d filas y %d columnas"%(counter,n_cols))
```

# Convirtiendo un diccionario en Data Frame

```python
df3 = pd.DataFrame(main_dict)
```

# Función para la lectura de datos desde una URL externa 

```python
def GetDataSetFromUrl(url,filename,download=False,
                           mainpath="/Users/erickfimbres/Documents/Erick Fimbres/Proyectos/Machine-Learning/datasets/",
                           type_file="csv",split=",",delim="\n",encoding="UTF-8",method="GET"):
    http = urllib3.PoolManager()
    
    r = http.request(method,url)
    r.status
    response = r.data
    #como la variable response viene en un tipo de datos string binario hay que decodificarlo a string mediante el metodo decode utf-8
    string_data = response.decode(encoding)

    #como el separador es el \n dividimos en un arreglo de filas
    lines = string_data.split(delim)

    col_names = lines[0].split(split)
    n_cols = len(col_names)

    data = {}
    counter = 0
    for name in col_names:
        data[name]=[]

    for line in lines[1:]:
        values = line.strip().split(split)
        for i in range(len(col_names)):
            data[col_names[i]].append(values[i])
        counter += 1

    #Convertimos a data frame
    data_frame = pd.DataFrame(data)
    
    #filename="olympics/download_medals."
    fullpath = os.path.join(mainpath,filename)
    
    if download:
        #Guardamos en el formato correspondiente
        if type_file == "csv":
            data_frame.to_csv(fullpath+"."+type_file)
        if type_file == "json":
            data_frame.to_json(fullpath+"."+type_file)
        if type_file == "xlsx" or type_file == "xls":  
            data_frame.to_excel(fullpath+"."+type_file)
        print("El archivo se ha guardado correctamente")
    return data_frame
```
